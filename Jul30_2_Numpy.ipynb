# Part2. 배열

Numpy의 배열은 모두 같은 자료형을 갖는 값이고, 음수가 아닌 정수의 튜플로 인덱싱

배열의 shape는 각 차원별 배열 크기의 tuple

중첩된 파이썬 리스트로부터 numpy 배열을 초기화 할 수 있고, 대괄호를 사용해서 요소로 접근 가능 O

```python
ex = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(ex) # list 속에 list
print(np.array(ex))
exx = np.array(ex)
exx.shape # shape로 확인함으로써 몇 개의 데이터가 있는지, 몇 차원으로 존재하는지 알 수 있음
```

```python
a = np.array([1, 2, 3])
print(a)

print(type(a), a.ndim, a.shape, a.dtype)

# type : 자료형
# ndim : 몇 차원인지
# shape : 튜플 형식이기 때문에, (1차원이라서) 뒤에가 비어있고, 콤마로 구분
# dtype : 배열 요소의 자료형

# [1, 2, 3]
# <class 'numpy.ndarray'> 1 (3,) int64
```

# Array BroadCasting

브로드캐스팅(BroadCasting)은 모양이 다른 배열들 간의 연산이 어떠한 조건이 갖춰졌을 때(만족했을 때) 가능해지도록 배열을 자동적으로 변환하는 것

- 요소가 하나인 배열인 경우는 어떤 배열이든 브로드캐스팅이 가능
- 하나의 배열이 1차원 배열인 경우 브로드캐스팅 가능
- (각 배열들이 다차원일 경우) 차원의 짝이 맞을 때, 브로드캐스팅이 가능

=> 서로 크기가 다른 array들의 연산이 가능해졌다!

```python
arr1
```

```python
arr3 = np.array([7, 8, 9])
arr3
```

```python
print(arr1.shape, arr3.shape)
```

```python
# arr1, arr3 사칙연산

# 덧셈
print(arr1 + arr3)
# 뺄셈
print(arr1 - arr3)
# 곱셈
print(arr1 * arr3)
# 나눗셈
print(arr1 / arr3)

# arr3의 array([7, 8, 9]) 가
# 어떤 조건에 만족해서
# [[7, 8, 9], [7, 8, 9]] 로 확장이 되어서 계산이 가능!
```

# Indexing과 Slicing의 차이

Indexing을 사용하면 항상 랭크(차원) 감소

Slicing을 사용하면 랭크(차원) 유지

```python
# 몇 차원인가...? 중괄호의 갯수로 파악하면!
b = np.arange(1, 13).reshape(3, 2, 2)
print(b)
print(b.ndim)
```

```python
indexing = b[1]
slicing = b[1:2]
print(indexing, indexing.shape, indexing.ndim)
print(slicing, slicing.shape, slicing.ndim)
```

```python
# array 수정

arr = np.array([11, 22, 33, 44, 55])
print(arr)

# 0번째 인덱스에 있는 값을 10으로 변경
arr[0] = 10
print(arr)
# 1, 2번째 인덱스에 있는 값을 222, 333 으로 변경
arr[1:3] = [222, 333]
print(arr)
# 3, 4번째 인덱스에 있는 값에 2를 더하기
arr[3:] += 2
print(arr)
```

```python
arr = np.array([110, 222, 337, 420, 155, 987, 200, 53, 178])
print(arr)
print(arr.shape)
# arr에서 200보다 큰 요소들에 대해서 기존 값에 2를 더한 값으로 수정
# 1.
for i, v in enumerate(arr):
    if v > 200:
        arr[i] += 2

# 2.
# for i in range(len(arr)): # len(arr) = 0, arr.shape[0]
#     arr[i] = arr[i] + 2 if arr[i] > 200 else arr[i]
print(arr)
```

```python
arr = np.array([110, 222, 337, 420, 155, 987, 200, 53, 178])
print(arr)

# 인덱스 안에서 조건을 걸 수 있음!!!
arr[arr > 200] += 2
print(arr)
```

```python
# array 삭제

arr = np.array([11, 22, 33, 44, 55])
print(arr)

# 0번째, 2번째 인덱스에 있는 값 삭제
# np.delete(array, index)
arr = np.delete(arr, [0, 2])
print(arr)
```

```python
# array 인덱스 추출
# np.where(조건)

# arr의 각 요소의 값이 100보다 작은 index 추출
np.where(arr < 100)
```

```python
arr = np.array([11, 130, 370, 114, 55, 94, 200, 30, 190, 187])

# np.where의 기능을 살펴보고,
# arr의 각 요소의 값이 100이상이면 'A', 아니면 'B'를 출력
print(arr)
print(np.where(arr >= 100, 'A', 'B'))
```
