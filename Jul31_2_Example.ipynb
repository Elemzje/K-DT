```python
# 1. 남자아이, 여자아이 출생 수 평균을 구해주세요 (정수로 출력)
answer1 = np.array(np.mean(birth, axis=0, dtype=np.int64)[1:])
print("남자아이", answer1[0])
print("여자아이", answer1[1])
# answer1
```

```python
#1 예답
print("남자아이 평균 : ", int(birth[:, 1].mean()))
print("여자아이 평균 : ", int(birth[:, 2].mean()))
```

```python
# 2. 2000년부터의 데이터를 구해주세요 (2000 ~ 2011)
#   연도    남자    여자
#   2000    1234567 1234567
#   ...
q2 = birth[np.where(birth[:, 0] >= 2000)]
print('연도', ' 남자', '    여자')
answer2 = []
for ans in q2:
    print(ans[0], ans[1], ans[2])
    answer2.append(ans)
answer2 = np.array(answer2)
# answer2
```

```python
#2 예답
print('연도', '남자', '여자', sep="\t")
print("=" * 30)
for y, m, f in birth[birth[:, 0] >= 2000]:
    print(y, m, f, sep="\t")
```

```python
# 3. 남자아이 출생이 가장 많은 연도와 남자아이 출생 수를 구해주세요
q3 = np.argmax(birth, axis=0)
boy = q3[1]
answer3 = np.array(birth[boy][:2])
print(answer3[0], answer3[1])
# answer3
```

```python
#3 예답
# argmax() : 가장 많은 값의 index를 반환
a = birth[birth[:, 1].argmax()]
y = a[0] # 연도
b = a[1] # 남아 출생수

print("연도", "남자아이", sep="\t")
print(y, b, sep='\t')
```

```python
# 4. 여자아이 출생이 가장 많은 top5를 구해주세요 (연도, 여자아이 수)
q4 = np.sort(birth, axis=0)
top5girl = q4[-5][2]
idx4 = np.where(birth[:, 2] >= top5girl)
arr4 = birth[idx4]
answer4 = []
for li in arr4:
    print(li[0], li[2])
    answer4.append([li[0], li[2]])
answer4 = np.array(answer4)
# answer4
```

```python
#4 예답

top5_girls = np.sort(birth[:, 2])[-1:-6:-1] # 역순으로 찾을때는 인덱스에 -1
print(top5_girls)

print("연도", "여자아이 수", sep="\t")
print("=" * 20)
for i in birth:
    for j in top5_girls:
        if (i[2] == j):
            print(i[0], j, sep='\t')
```

```python
# 5. 여자아이 출생률(%)을 구하고, 각 연도별로 여자아이 출생수가 차지하는 비율을
#   데이터로 보여주세요. (소수점 둘째자리까지)
#   연도    남자    여자    여자출생률
#   1880    123123  100000  41
#   ...
sums = np.sum(birth, axis=0)
girlbirthrate = sums[2] / (sums[1] + sums[2])
print("%.2f%%" % (girlbirthrate * 100))
print('연도  남자  여자  여자출생률')
answer5 = []
for li in birth:
    bornrate = (li[2] / (li[2] + li[1])) * 100
    print(li[0], li[1], li[2], bornrate.round(2))
    answer5.append([li[0], li[1], li[2], bornrate])
answer5 = np.array(answer5, dtype=np.int64)
# answer5
```

```python
#5 예답
girls_ratio = ((birth[:, 2] / (birth[:, 1] + birth[:, 2])) * 100).round(2)
girls_ratio

print('연도', '남아', '여아', '여아 출생률', sep='\t')
print('=' * 40)
for y, b, g, gr in zip(birth[:, 0], birth[:, 1], birth[:, 2], girls_ratio):
    print(y, b, g, gr, sep="\t")
```

```python
# 6. 2000년 이후 데이터에 대해 출생량을 구하고,
#   남자아이 수가 1800000 이상이면 '많음', 아니면 '적음' 이라는 내용을 출력
after2000 = answer2
answer6 = np.where(after2000[:, 1] >= 1800000, "많음", "적음")
for i in range(len(answer6)):
    print(after2000[i][0], answer6[i])
# answer6
```

```python
#6 예답
after_2000 = birth[birth[:, 0] >= 2000]
# print(after_2000)

boys = after_2000[:, 1]
# print(boys)

amount = np.where(boys >= 1800000, "많음", "적음")
# print(amount)

for y, b, amt in zip(after_2000[:, 0], boys, amount):
    print(y, b, amt)
```

```python
# 7. 1800년대, 1900년대, 2000년대 각 남자아이, 여자아이 출생수의 평균을 구해주세요
before1900 = birth[np.where(birth[:, 0] < 1900)]
twcen = birth[20:-12]
after2000
cs = [before1900, twcen, after2000]
answer7 = []
for c in cs:
    line = np.round(np.mean(c, axis=0))
    print(line[1], line[2])
    answer7.append([line[1], line[2]])
answer7 = np.array(answer7)
# answer7
```

```python
#7 예답
print("1800년대")
print("남아 평균", "여아 평균")
k_1800s = birth[ (birth[:, 0] >= 1800) & (birth[:, 0] < 1900) ][:, 1:]
# print(k_1800s)
print(k_1800s.mean(axis=0)[0], k_1800s.mean(axis=0)[1])

print()

print("1900년대")
print("남아 평균", "여아 평균")
k_1900s = birth[ (birth[:, 0] >= 1900) & (birth[:, 0] < 2000) ][:, 1:]
# print(k_1900s)
print(k_1900s.mean(axis=0)[0], k_1900s.mean(axis=0)[1])

print()

print("2000년대")
print("남아 평균", "여아 평균")
k_2000s = birth[ (birth[:, 0] >= 2000) & (birth[:, 0] < 2100) ][:, 1:]
# print(k_2000s)
print(k_2000s.mean(axis=0)[0].round(2), k_2000s.mean(axis=0)[1])
```

```python
# 8. 여자아이 출생수 25% ~ 75% 사이의 값을 구해주세요 (연도, 여자아이 출생 수)
startend = np.quantile(birth[:, 2], [0.25, 0.75])
girlsort = np.sort(birth, axis=0)
startidx = np.where(girlsort[:, 2] > startend[0])[0][0]
endidx = np.where(girlsort[:, 2] > startend[1])[0][0]
answer8 = []
for i in range(startidx, endidx + 1):
    year = np.where(birth[:, 2] == girlsort[i, 2])
    print(birth[year][0][0], girlsort[i, 2])
    answer8.append([birth[year][0][0], girlsort[i, 2]])
answer8 = np.array(answer8)
# answer8
```
