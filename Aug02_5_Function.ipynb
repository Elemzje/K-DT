# 함수 매핑(mapping)

시리즈나 데이터프레임의 개별 요소를 특정 함수에 1:1로 대응시키는 과정

사용자가 직접 만든 함수(람다 포함)를 적용시킬수 있기 때문에 for문이나 기본 함수로 처리하는 것보다 훨씬 효율적인 장점

```python
# seaborn에서 타이타닉 데이터 불러와서 'age', 'fare'열만 인덱싱
#   => 'ten' 열을 만들어서 => 10을 넣기
t = sns.load_dataset('titanic')

df = t.loc[:, ['age', 'fare']]
df['ten'] = 10
df.head()
```

```python
# 값을 넣으면 10을 더해주는 함수
def add_10(a):
    return a + 10

# 값을 2개 넣으면 그 둘을 곱해주는 함수
def mul_two_obj(a, b):
    return a * b

# 값을 3개 넣으면 그 셋을 더해주는 함수
def add_three_obj(a, b, c):
    return a + b + c

print(add_10(15))
print(mul_two_obj(10, 20))
print(add_three_obj(10, 20, 30))
```

```python
# 이 함수들을 apply()를 이용해서 시리즈객체의 개별요소에 적용
sr1 = df['age'].apply(add_10) # age의 모든 요소 + 10
sr1.head()
```

```python
sr2 = df['fare'].apply(mul_two_obj, b=0.9)
sr2.head()
# apply함수의 첫번째 파라미터 : 함수
# 두번째 파라미터 : 값
```

```python
sr3 = df['fare'].apply(add_three_obj, b=10.1, c=20)
sr3.head()
```

```python
# 'ten' 열의 값들을 lambda함수를 이용해서 add_10함수 적용
# (lambda a, b: print(a * b))(10, 20)
sr4 = df['ten'].apply(lambda x: add_10(x))
sr4.head()
```

# apply() vs applymap()

```python
df_map = df.applymap(add_10) # 데이터프레임의 모든 요소에 함수를 적용함
df_map.head()
```

# 시리즈 객체 함수 매핑

DataFrame객체.apply(매핑함수, axis=값)

열 or 행에 대해서 함수를 적용시킬수 있음

```python
# 뭔가를 넣었을 때 결측값을 확인하는 함수
def missing_value(series):
    return series.isnull()
```

```python
# 이 함수를 데이터프레임의 모든 열에 매핑 (axis = 1)
result = df.apply(missing_value, axis=1)
print(result)
print()
result.describe()
```
