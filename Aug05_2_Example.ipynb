```python
# 필요한 라이브러리 불러오기
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# http://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=ss209
# 전체 내용 데이터를 크롤링하기
# parser 종류 : html.parser => html5lib
u = 'http://www.menupan.com/restaurant/bestrest/bestrest.asp?pt=rt&areacode=ss209'
res = requests.get(u)
# res.status_code
if res.status_code == 200:
    html = res.text
    # print(res.content)
    # print(html)
```

```python
# 순위, 상호, 업종, 주소, url에 대한 내용을 출력

# select() vs select_one()
# select() : 매칭되는 여러 값들을 리스트에 담음
# select_one() : 매칭되는 값 하나만 가져와서 => 객체

# select관련 => 하나의 열에 여러 선택자가 필요한 경우
#   => .OOO or .XXX => (or : ',' (쉼표))

# 속성값 구해야 할때 : [ ]를 사용!!

soup = BeautifulSoup(html, 'html5lib') # html.parser, xml, lxml, html5lib
ul = soup.select_one('div.rankingList > ul.list')
lists = ul.select('li')
# print(lists)
for li in lists:
    rank = li.select_one('.rankNum, .numTop').text
    name = li.select_one('span.restName > a').text
    menuType = li.select_one('p.listType').text
    address = li.select_one('p.listArea').text
    url = li.select_one('p > a').attrs['href']
    print(rank, name, menuType, address, url)
    print(type(li.select_one('p > a')))
```

```python
# 위의 데이터들을 list로 나타내는 작업

# [ {'순위' : '1', '상호': 'XXXX', '업종':'한식', ...},
#   {'순위' : '2', '상호': 'XXXX', '업종':'한식', ...},
#   {'순위' : '3', '상호': 'XXXX', '업종':'한식', ...},
# ... ]

# url에 대한 내용은 따로 별개의 list에 넣기

menuList = []
urlList = []

for li in lists:
    listdict = {}
    listdict['순위'] = li.select_one('p').text
    listdict['상호'] = li.select_one('span.restName > a').text
    listdict['업종'] = li.select_one('p.listType').text
    listdict['주소'] = li.select_one('p.listArea').text
    menuList.append(listdict)
    urlList.append(li.select_one('p > a').attrs["href"])
print(menuList)
print(urlList)
```

```python
# menuList를 가지고 순위를 index로 하는 DataFrame을 생성
df = pd.DataFrame(menuList)
df.set_index('순위', inplace=True)
df.head()
```

```python
# http://www.menupan.com/이 기본 주소값 뒤에 urlList에 담아놓은 주소값 붙여서 요청
# 전화번호, 상세주소(두개 중 하나)를 가져온 후에
defaultUrl = 'http://www.menupan.com'
focusList = []
for urlli in urlList:
    url = defaultUrl + urlli
    # print(url)
    res = requests.get(url)
    if res.status_code == 200:
        soup = BeautifulSoup(res.text, 'html5lib')
        arBasic = soup.select_one('div.areaBasic')
        tel = arBasic.select_one('dl.restTel > dd').text
        add = arBasic.select_one('dl.restAdd > dd.add1').text
        lidict = {}
        lidict['전화번호'] = tel
        lidict['상세주소'] = add
        focusList.append(lidict)
print(focusList)
```

```python
# 만들어놓은 DataFrame에 붙이기 ( + '주소' 열은 삭제)
try:
    df.drop('주소', axis=1, inplace=True)
except:
    print('이미 삭제됨')
focus = pd.DataFrame(focusList)
df['전화번호'] = list(focus['전화번호'])
df['상세주소'] = list(focus['상세주소'])
df.head()
```

```python
# 업종이 한식인 가게의 정보만 출력
# df.where(df['업종'] == '한식').dropna()
df[df['업종'] == '한식']
```
